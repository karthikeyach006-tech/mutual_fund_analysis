```python
import pandas as pd

nav_history = pd.read_csv("../data/raw/02_nav_history.csv")


nav_history["date"] = pd.to_datetime(nav_history["date"])

```

```python
nav_history = nav_history.sort_values(
    ["amfi_code", "date"]
)
```

```python
nav_history = nav_history.drop_duplicates()
```

```python
nav_history["nav"] = (
    nav_history.groupby("amfi_code")["nav"]
    .ffill()
)
```

```python
invalid_nav = nav_history[
    nav_history["nav"] <= 0
]

print("Invalid NAV rows:", len(invalid_nav))
```

```python
nav_history.to_csv(
    "../data/processed/nav_history_clean.csv",
    index=False
)

print("Saved successfully!")
```

```python
performance = pd.read_csv(
    "../data/raw/07_scheme_performance.csv"
)
```

```python
return_cols = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct"
]

for col in return_cols:
    performance[col] = pd.to_numeric(
        performance[col],
        errors="coerce"
    )
```

```python
performance[return_cols].isnull().sum()
```

```python
anomalies = performance[
    (performance["return_1yr_pct"] > 100) |
    (performance["return_1yr_pct"] < -100)
]

print("Anomalies:", len(anomalies))
```

```python
expense_issues = performance[
    (performance["expense_ratio_pct"] < 0.1) |
    (performance["expense_ratio_pct"] > 2.5)
]

print(
    "Expense Ratio Issues:",
    len(expense_issues)
)
```

```python
performance = performance.drop_duplicates()
```

```python
performance.to_csv(
    "../data/processed/scheme_performance_clean.csv",
    index=False
)
```
